# Visualization

```python
# spliting data in numeric and text
colnumber = len(data.columns)
data_number = pd.DataFrame()
data_text = pd.DataFrame()
for i in range(colnumber):
    if data.dtypes[i] == "int64":
        data_number[[data.columns[i]]] = pd.DataFrame(data.iloc[:,i])
    else:
        data_text[[data.columns[i]]] = pd.DataFrame(data.iloc[:,i])

data_text["Total Household Income"] = data.iloc[:,0] # add "Total Household Income" to this data
```

```python
income = data["Total Household Income"]
income.hist(bins=120, figsize=(19,10), color = "g")
plt.show()
```

```python
expenditures = data_number.iloc[:,0:22].drop("Agricultural Household indicator", axis = 1)
```

```python
expenditures.hist(bins=50, figsize=(20,16))
plt.show()
```
